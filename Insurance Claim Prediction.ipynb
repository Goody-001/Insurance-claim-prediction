```python
data['Date_of_Occupancy'].mean() 
median_occupancy_year = data['Date_of_Occupancy'].median()
```

```python
data['Date_of_Occupancy_Imputed'] = data['Date_of_Occupancy'].fillna(median_occupancy_year)
data['Building_Age'] = data['YearOfObservation'] - data['Date_of_Occupancy_Imputed']
```

```python
X = data.drop(columns=['Claim'])
y = data['Claim']
```

```python
numeric_features = X.select_dtypes(include=['int64','float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
numeric_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer([
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
pipe_lr = Pipeline([
    ('preprocessor', preprocessor),
    ('model', LogisticRegression(max_iter=1000, class_weight='balanced'))
])

pipe_rf = Pipeline([
    ('preprocessor', preprocessor),
    ('model', RandomForestClassifier(random_state=42))
])

pipe_gb = Pipeline([
    ('preprocessor', preprocessor),
    ('model', GradientBoostingClassifier(random_state=42))
])

models = {
    'Logistic Regression': pipe_lr,
    'Random Forest': pipe_rf,
    'Gradient Boosting': pipe_gb
}

for model in models.values():
    model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score, classification_report

for name, model in models.items():
    preds = model.predict_proba(X_test)[:,1]
    print(name, roc_auc_score(y_test, preds))
```

```python
param_rf = {
    'model__n_estimators': [200, 300],
    'model__max_depth': [None, 10]
}

grid_rf = GridSearchCV(pipe_rf, param_rf, scoring='roc_auc', cv=5)
grid_rf.fit(X_train, y_train)
```

```python
param_gb = {
    'model__n_estimators': [100, 200],
    'model__learning_rate': [0.05, 0.1]
}

grid_gb = GridSearchCV(pipe_gb, param_gb, scoring='roc_auc', cv=5)
grid_gb.fit(X_train, y_train)
```

```python
y_rf = grid_rf.predict_proba(X_test)[:,1]
y_gb = grid_gb.predict_proba(X_test)[:,1]
```

```python
print("Tuned Random Forest ROC-AUC:",
      roc_auc_score(y_test, y_rf))

print("Tuned Gradient Boosting ROC-AUC:",
      roc_auc_score(y_test, y_gb))
```

```python
tuned_models = {"Random Forest Classifier": y_rf, "Gradient Boosting": y_gb}

for name, y_proba in tuned_models.items():
    y_pred = (y_proba >= 0.5).astype(int)
    print(name)
    print("ROC-AUC:", roc_auc_score(y_test, y_proba))
    print(classification_report(y_test, y_pred))
```
